## DataFrame (Tablas Mejoradas)

```python
# Creo 2 Listas
lista1 = [20, 21, 35, 45]
lista2 = ['Rodrigo', 'Maximiliano', 'Raul', 'Italo']
```

```python
# Genero la Estructura de la Tabla
estruc_tabla = {'edad' : lista1, 'nombre' : lista2}
```

```python
# Genero el DataFrame desde la Estructura
dataframe = pd.DataFrame(estruc_tabla)
```

```python
# Muestro el DataFrame
dataframe
```

```python
dataframe.describe()
```

```python
# extraer 1ra columna

# 1ra opcion
# dataframe.edad

#2da opcion (recomendada si el nombre de la columna tiene espacios ' ')
dataframe['edad']
```

```python
# Métodos de un DataFrame
# min   -   max -   median  -   std -   mode    -   sum -   count   -   unique  -   nunique -   mean

# Mínimo
print( dataframe['edad'].min() )

# Máximo
print( dataframe['edad'].max() )

# Desviación Estandar
print( dataframe['edad'].std() )

print( dataframe['edad'].mode() )

# Suma de todos los Valores
print( dataframe['edad'].sum() )

# Cantidad de Valores
print( dataframe['edad'].count() )

# Mediana
print( dataframe['edad'].median() )

# Lista de Valores Únicos
print( dataframe['edad'].unique() )

# Cantidad de Valores Únicos
print( dataframe['edad'].nunique() )

# Promedio
print( dataframe['edad'].mean() )
```

```python
# Obtener 3er dato
dataframe.loc[2]
```

```python
# Obtener 2do y 3er dato
dataframe.loc[1:2]
```

```python
# Obtener 1er y 3er dato (paso como parámetro una lista de los valores que quiero ver)
dataframe.loc[[0,2]]
```

```python
# Método iloc no importa como se llamen los indices los toma como valores numericos 0 --> n-1
dataframe.iloc[1:4]
```

```python
# Agrego una nueva columna
dataframe['edad_al_cuadrado'] = dataframe['edad'] ** 2
dataframe
```

```python
# Muestro solamente las columnas de edades (paso como parámetros una lista con las columnas que quiero ver)
dataframe[['edad', 'edad_al_cuadrado']]
```

```python
# Muestro las primeras 'n' filas
dataframe.head(2)
```

```python
# Muestro las últimas 'n' filas
dataframe.tail(2)
```

```python
# Muestra aleatoria de 'n' filas
dataframe.sample(2)
```

```python
# Filtrar por condicion
# dataframe [   condicion   ]
dataframe[  dataframe['edad'] > 30  ]
```

```python
# Filtrar por condicion
# dataframe [   cond1  &   cond2   ]
dataframe[  (dataframe['edad'] < 25) | (dataframe['nombre'] == 'Italo') ]
```

```python
# Eliminar Filas o Columnas
# axis 0 --> Filas
# axis 1 --> Columnas
dataframe = dataframe.drop('edad_al_cuadrado', axis=1)
dataframe
```

Ejemplo

```python
df_lluvias = pd.read_csv('pune_1965_to_2002.csv')
```

```python
df_lluvias
```

```python
# Reemplazo los índices de la tabla por los 'Años' que estan en la 1er columna
df_lluvias.index = df_lluvias['Year']
# Elimino la 1er columna ya que ahora son los índices
df_lluvias = df_lluvias.drop('Year', axis='columns')
```

```python
df_lluvias.sample(3)
```

```python
# Armo los ejes
x = df_lluvias.index
y = df_lluvias.sum(axis='columns')
```

```python
# Método de codificación como Objetos

    # Defino el Lienzo
fig, ax = plt.subplots(figsize = (12, 4))   # figsize --> tamaño gráfico (ancho, alto)

    # Configuración del Gráfico
        # Etiquetar ejes
ax.set_xlabel('Años')
ax.set_ylabel('Precipitación acumulada en mm.')
        
        # Añadir título
ax.set_title('Precipitaciones acumuladas segun el año')
        
        # Definir Límites
ax.set_xlim(df_lluvias.index[0], df_lluvias.index[-1])
            # ax.set_ylim
        
        # Líneas [h]orizontales y [v]erticales
            # ax.axhline
            # ax.axvline
    
    # Ejecución de la/s Gráfica/s
        # ax.plot --> gráfico de líneas
        # ax.scatter --> gráfico de puntos
        # ax.bar --> gráfico de barras
ax.plot(x, y, label = 'Precipitaciones acumuladas')

        # Añadir leyenda
ax.legend(loc = 'best')    # localización --> best, right, etc.
```

```python
# Exportar la imagen, formatos soportados: jpeg, jpg, png, pdf, svg
fig.savefig('precipitaciones_anio.pdf')
fig.savefig('precipitaciones_anio.jpg')
```
